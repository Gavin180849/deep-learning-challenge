# Analysis for Deep Learning model for Nonprofit Foundation Alphabet Soup
### Overview:
The purpose of this analysis is to create and optimize a deep learning neural network to help the nonprofit foundation, Alphabet Soup, select applicants requesting funding who will be successful at using funds effectively.

## Results:

### Data Preprocessing:

**What variable(s) are the target(s) for your model?**
The target for the module was the column "IS_SUCCESSFUL".

**What variable(s) are the features for your model?**
The feature variables were the "APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "STATUS", "INCOME_AMT", "SPECIAL_CONSIDERATIONS", AND "ASK_AMT". The "IS_SUCCESSFUL" column was dropped from the dataframe when preprocessing.

**What variable(s) should be removed from the input data because they are neither targets nor features?**
Both the "EIN" and "NAME" columns should be removed because they are identification columns and not needed to select future applicants.

## Compiling, Training, and Evaluating the Model

**How many neurons, layers, and activation functions did you select for your neural network model, and why?**
I used 2 neurons per feature, 2-3 hidden layers and I utilized the relu function for the hidden layers due to the complexity. Sigmoid activation function was used for the outer layer. These were used to try to reach the 75% accuracy.

**Were you able to achieve the target model performance?**
No, I only had 1 Epoch where the model reached 75% on the first run.

**What steps did you take in your attempts to increase model performance?**
The second attempt I increasesd the hidden layers and the nodes, I also decreased the epochs. In the third attempt I increased the nodes and kept the three hidden layers, epochs were the same as attempt 2.
